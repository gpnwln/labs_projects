## Задание 4 (Task 4)
#### Рассчитаем главные компоненты и возьмем первые по методу Кайзера. (Let's calculate the main components and take the first ones according to the Kaiser method.)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Возьмем, например, 11 главных компонент = количество признаков.(Take, for example, 11 main components = the number of features.)

```python
pca = PCA(n_components = 11)
table_pca = pca.fit_transform(df[Columns])
explained_variance_ratio = pca.explained_variance_ratio_
print("Дисперсии компонент (vector of the variance): \n", explained_variance_ratio)
explained_variance = pca.explained_variance_
print("Собственные значения (eigenvalues): \n", explained_variance)
```

#### Критерий Кайзера (Kaiser's Criterion)

```python
pc = 0
for i in range(0, 10):
    if (explained_variance[i] >= 1):
        pc += 1
    else:
        break
        
print(f'Значения больше 1 имеют {pc} главных компонент')
```

#### По критерию Кайзера получаем 4 главные компоненты, но у пятой главной компоненты собственное число равно примерно 0.96, поэтому возьмем 5 главных компонент. (According to the Kaiser criterion, we get 4 main components, but the fifth main component has an eigenvalue of about 0.96, so we take 5 main components.)

```python
# pc = 5
```

#### Корреляционная матрица в виде тепловой карты (Correlation matrix in the form of a heat map)

```python
correlation = df[Columns].corr()
plt.figure(figsize=(11,11)) 
sns.heatmap(correlation, vmax=1, square=True, annot=True,cmap='Blues')
plt.title('Correlation between different features')
plt.show()
```

```python
eigenvalues, eigenvector = np.linalg.eig(correlation)
```

#### Коэффициенты линейного преобразования центрировано-нормированных исходных признаков (Coefficients of linear transformation of centered-normalized initial features)

```python
vector_df = pd.DataFrame(eigenvector, columns=['Factor_' + str(i + 1) for i in range(len(Columns))], index=Columns)
```

#### Матрица нагрузок (Load matrix)

```python
A = vector_df @ np.diag(np.sqrt(explained_variance))
A.index = Columns
```

#### Матрица индивидуальных значений главных компонент (Matrix of individual values of the main components)

```python
F = (np.linalg.inv(A.T @ A) @ A.T @ df[Columns].T).T
F.columns = ['Factor_' + str(i + 1) for i in range(len(Columns))]
F = F.iloc[:, :pc]
F['Class'] = df['Class']
F
```

```python
F['color'] = df['Class']
F['color'] = F['color'].apply(lambda x: colors_map[x])
x = F.iloc[:, 0]
y = F.iloc[:, 1]
plt.title("Real after PCA", fontweight='bold')
plt.xlabel(titles[0])
plt.ylabel(titles[1])
plt.scatter(x, y, c = F['color'], s = 10)
```

## Задание 5 (Task 5)
#### Выполним кластеризацию данных с помощью метода Affinity Propagation и сравниваем число кластеров с реальным. (Let's perform data clustering using the Affinity Propagation method and compare the number of clusters with the real one.)

```python
from sklearn.cluster import AffinityPropagation
```

```python
print("Number of real classes = ", len(df['Class'].unique()))
```

```python
af = AffinityPropagation().fit(F[['Factor_1','Factor_2','Factor_3','Factor_4']])
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
print("Центры класстеров (Cluster centers): ", cluster_centers_indices)
print("Количество класстеров (Number of clusters): ",len(cluster_centers_indices))
```

#### Без подбора параметров данный метод показывает разбиение для 80 классов, поэтому подберем параметры, чтобы было 6 классов. (Without selecting parameters, this method shows a partition for 80 classes, so we will select the parameters so that there are 6 classes.)

```python
af = AffinityPropagation(preference = -180, damping = 0.9).fit(F[['Factor_1','Factor_2','Factor_3','Factor_4']])
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
print("Центры класстеров (Cluster centers): ", cluster_centers_indices)
print("Количество класстеров (Number of clusters): ",len(cluster_centers_indices))
```

```python
F['color_af'] = labels
F['color_af'] = F['color_af'].apply(lambda x: colors_map[x])
```

```python
plt.title("AP", fontweight='bold')
plt.xlabel(titles[0])
plt.ylabel(titles[1])
plt.scatter(x, y, c = F['color_af'], s = 10)

plt.scatter(F.iloc[cluster_centers_indices[0], 0],
F.iloc[cluster_centers_indices[0], 1], c = ['lime'], s = 70, marker = 'X')

plt.scatter(F.iloc[cluster_centers_indices[1], 0],
F.iloc[cluster_centers_indices[1], 1], c = ['lime'], s = 70, marker = 'X')

plt.scatter(F.iloc[cluster_centers_indices[2], 0],
F.iloc[cluster_centers_indices[2], 1], c = ['lime'], s = 70, marker = 'X')

plt.scatter(F.iloc[cluster_centers_indices[3], 0],
F.iloc[cluster_centers_indices[3], 1], c = ['lime'], s = 70, marker = 'X')

plt.scatter(F.iloc[cluster_centers_indices[4], 0],
F.iloc[cluster_centers_indices[4], 1], c = ['lime'], s = 70, marker = 'X')

plt.scatter(F.iloc[cluster_centers_indices[5], 0],
F.iloc[cluster_centers_indices[5], 1], c = ['lime'], s = 70, marker = 'X')

plt.show()
```

## Задание 6 (Task 6)
### Используя метод силуэта, определяется оптимальное число кластеров для кластеризации методами DBSCAN и Spectral Clustering и сравниваем с реальным числом кластеров. (Using the silhouette method, the optimal number of clusters for clustering is determined by DBSCAN and Spectral Clustering methods and compared with the real number of clusters.)

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import SpectralClustering
```

```python
print("Number of real classes = ", len(df['Class'].unique()))
```

```python
silhouette = []
max = 0
rg = range(2, 11)
for k in rg:
    sc1 = SpectralClustering(n_clusters = k, affinity = 'nearest_neighbors').fit(F[['Factor_1','Factor_2','Factor_3','Factor_4']])
    preds_labels_sc1 = sc1.labels_
    silhouette.append(silhouette_score(F[['Factor_1','Factor_2','Factor_3','Factor_4']],preds_labels_sc1))
```

```python
plt.title("Метод силуэта\n",fontsize=16)
plt.scatter(x=[i for i in rg],y=silhouette,s=150,edgecolor='k')
plt.plot(rg, silhouette)
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Силуэт",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

```python
silhouette = []
max = 0
rg = range(2, 11)
for k in rg:
    sc2 = SpectralClustering(n_clusters = k, affinity = 'rbf').fit(F[['Factor_1','Factor_2','Factor_3','Factor_4']])
    preds_labels_sc2 = sc2.labels_
    silhouette.append(silhouette_score(F[['Factor_1','Factor_2','Factor_3','Factor_4']],preds_labels_sc2))
```

```python
plt.title("Метод силуэта\n",fontsize=16)
plt.scatter(x=[i for i in rg],y=silhouette,s=150,edgecolor='k')
plt.plot(rg, silhouette)
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Силуэт",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

#### -метода ближайших соседей равно 5. (the nearest neighbor method is 5.)
