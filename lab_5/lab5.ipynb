### Соответсвенно, наилучшим MLP-классификатором является следующий с score = 46,67%:
#### MLPClassifier(alpha=1e-06, hidden_layer_sizes=(4, 4), learning_rate_init=0.0001, random_state=22222, solver='lbfgs'), 'lbfgs', 1e-06, 'constant', 0.0001, 4, 4]

```python
clf = MLPClassifier(solver = 'lbfgs', alpha = 1e-06, learning_rate = 'constant', learning_rate_init = 0.0001, hidden_layer_sizes = (4, 4))
clf.fit(df_train_table[Columns], df_train_table['Class'])
res = clf.predict(df_test_table[Columns])
preds = clf.predict_proba(df_test_table[Columns])

df_test_table['mlp_class'] = res
df_test_table
```

```python
print(classification_report(df_test_table['Class'],df_test_table['mlp_class']))
print("ROC_AUC(macro) = ", roc_auc_score(df_test_table['Class'], preds, multi_class='ovr'))
```

## Задание 7 (Task 7)
### SVM-классификатор

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
```

```python
gammas = ['scale', 'auto', 1, 10, 0.1, 20, 0.01]
coef0s = [0, 1, 2, 5, 10, 0.1, 0.01]
degrees = [2, 3, 4, 5]
Cs = [0.001, 0.01, 0.1, 0.5, 1, 2, 10, 20]
```

```python
krl ='linear'

best = [0]
for gamma in gammas:
    for coef0 in coef0s:
        for degree in degrees:
            for C in Cs:
                clf = svm.SVC(kernel = krl, gamma = gamma, coef0 = coef0,
                              degree = degree, C = C)
                clf.fit(df_train_table[Columns], df_train_table['Class'])
                # res = clf.predict(df_test_table[Columns])
                # preds = clf.predict_proba(df_test_table[Columns])
                score = clf.score(df_test_table[Columns], df_test_table['Class'])
                if score > best[0]:
                    best = [score, gamma, coef0, degree, C]
                        
print(krl, best)

clf = svm.SVC(kernel = krl, gamma = best[1], coef0 = best[2], degree = best[3], C = best[4])
clf.fit(df_train_table[Columns], df_train_table['Class'])
res = clf.predict(test_table[Columns])

print("Score = ", clf.score(df_test_table[Columns], df_test_table['Class']))
print("Accuracy = ", accuracy_score(df_test_table['Class'], res))
print("Precision = ", precision_score(df_test_table['Class'], res, average='macro', zero_division=1))
print("Recall = ", recall_score(df_test_table['Class'], res, average='macro'))
```
