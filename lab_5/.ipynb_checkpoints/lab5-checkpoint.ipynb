## Задания 1-2 (Tasks 1-2)
#### Загрузка и анализ исходных данных (Downloading and analyzing input data)

```python
import pandas as pd
```

```python
df = pd.read_csv('vine.dat', header=None)
df.columns =['FixedAcidity', 'VolatileAcidity', 'CitricAcid', 'ResidualSugar', 'Chlorides', 'FreeSulfurDioxide', 'TotalSulfurDioxide', 'Density', 'PH', 'Sulphates', 'Alcohol', 'Class']
df
```

```python
print(df['Class'].unique())
```

```python
df['Class'] = df['Class'] - 3
df.dtypes
```

-----------------
## Задание 3 (Task 3)
#### Стандартизация (Standardization)

```python
from sklearn import preprocessing
import numpy as np
```

#### Объявим переменную с названиями всех колонок кроме 'Class' (Declare a variable with the names of all columns except 'Class')

```python
Columns = ['FixedAcidity', 'VolatileAcidity', 'CitricAcid', 'ResidualSugar', 'Chlorides', 'FreeSulfurDioxide', 'TotalSulfurDioxide', 'Density', 'PH', 'Sulphates', 'Alcohol']
```

```python
scaler_std = preprocessing.StandardScaler()
df[Columns] = scaler_std.fit_transform(df[Columns])
df.dropna()
df
```

#### Данные стандартизированы и очищены от нулевых значений. (The data is standardized and cleared of zero values.)

-------------------
## Задание 4 (Task 4)
### Разделяем полученный датафрейм на 3 выборки: обучающую, тестовую и валидационную выборки в соотношении 5:3:2 соответственно.

```python
col = ['index'] + Columns + ['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_table, test_valid_table = train_test_split(df, train_size = 0.5)

train_table = train_table.reset_index()
test_valid_table = test_valid_table.reset_index()

df_train_table = pd.DataFrame(train_table, columns = col)
df_test_valid_table = pd.DataFrame(test_valid_table, columns = col)

print('train_table:')
df_train_table
```

#### Тестовая и валидационная выборки находятся в test_valid_table, соответственно тестовая и валидационная относятся друг к другу как 0,6 и 0,4 соответственно.

```python
test_table, valid_table = train_test_split(df_test_valid_table, train_size = 0.6)

test_table = test_table.reset_index()
valid_table = valid_table.reset_index()

df_test_table = pd.DataFrame(test_table, columns = col)
df_valid_table = pd.DataFrame(valid_table, columns = col)

print('test_table:')
df_test_table
```

```python
print('valid_table:')
df_valid_table
```

### Получаем 3 выборки в датафреймах: тестовую, обучающую и валидационную.

## Задание 5 (Task 5)

#### На основе обучающей и тестовой выборки постройте MLP-классификатор. Рассчитайте параметры его эффективности (Accuracy, Precision. Recall, ROC-AUC). Меняя значение параметров классификатора обосновано подберите наиболее удачную MLP-сеть

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(solver = 'lbfgs', alpha = 0.1, hidden_layer_sizes = (2, 2), random_state = 22222)
clf.fit(df_train_table[Columns], df_train_table['Class'])
res = clf.predict(df_test_table[Columns])

df_test_table['mlp_class'] = res
df_test_table
```
